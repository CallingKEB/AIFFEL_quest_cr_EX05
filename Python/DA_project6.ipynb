# 6-1. 프로젝트 (1) load_digits : 손글씨를 분류해 봅시다

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

import numpy as np
import pandas as pd
```

```python
# load_digits메서드를 사용하여 데이터 준비

digits = load_digits()
```

```python
# Feature Data 지정하기
X = digits.data
X

# Label Data 지정하기
Y = digits.target
Y

# Target Names 출력해 보기
Y_name = digits.target_names
Y_name
```

```python
# 데이터 Describe 해 보기
digits_df = pd.DataFrame(data=digits.data, columns=digits.feature_names)
digits_df['label'] = digits.target
digits_df.describe()
```

```python
# train, test 데이터 분리

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=23)
```

```python
# 5. 다양한 모델로 학습 및 6. 평
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Decision Tree
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, Y_train)

Y_pred_dt = decision_tree.predict(X_test)
print("Decision Tree 성능 평가")
print(classification_report(Y_test, Y_pred_dt))
print(f"Accuracy: {accuracy_score(Y_test, Y_pred_dt)}\n")
```

```python
# Random Forest
random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, Y_train)

Y_pred_rf = random_forest.predict(X_test)
print("Random Forest 성능 평가")
print(classification_report(Y_test, Y_pred_rf))
print(f"Accuracy: {accuracy_score(Y_test, Y_pred_rf)}\n")
```

```python
# Support Vector Machine (SVM)
svm = SVC(random_state=42)
svm.fit(X_train, Y_train)

Y_pred_svm = svm.predict(X_test)
print("SVM 성능 평가")
print(classification_report(Y_test, Y_pred_svm))
print(f"Accuracy: {accuracy_score(Y_test, Y_pred_svm)}\n")
```

```python
# SGD Classifier
sgd = SGDClassifier(random_state=42)
sgd.fit(X_train, Y_train)

Y_pred_sgd = sgd.predict(X_test)
print("SGD Classifier 성능 평가")
print(classification_report(Y_test, Y_pred_sgd))
print(f"Accuracy: {accuracy_score(Y_test, Y_pred_sgd)}\n")
```

```python
# Logistic Regression
logistic_regression = LogisticRegression(max_iter=10000, random_state=42)
logistic_regression.fit(X_train, Y_train)

Y_pred_lr = logistic_regression.predict(X_test)
print("Logistic Regression 성능 평가")
print(classification_report(Y_test, Y_pred_lr))
print(f"Accuracy: {accuracy_score(Y_test, Y_pred_lr)}\n")
```
